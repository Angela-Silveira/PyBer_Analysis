# Pyber Challenge

### 4.3 Loading and Reading CSV files

```python
# Add Matplotlib inline magic command
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd

# File to Load (Remember to change these)
city_data_to_load = "Resources/city_data.csv"
ride_data_to_load = "Resources/ride_data.csv"

# Read the City and Ride Data
city_data_df = pd.read_csv(city_data_to_load)
ride_data_df = pd.read_csv(ride_data_to_load)
```

### Merge the DataFrames

```python
# Combine the data into a single dataset
pyber_data_df = pd.merge(ride_data_df, city_data_df, how="left", on=["city", "city"])

# Display the data table for preview
pyber_data_df.head()
```

## Deliverable 1: Get a Summary DataFrame 

```python
#  1. Get the total rides for each city type
city_total_rides = pyber_data_df.groupby(["type"]).count()["ride_id"]
city_total_rides
```

```python
# 2. Get the total drivers for each city type
city_total_drivers = city_data_df.groupby(["type"]).sum()["driver_count"]
city_total_drivers
```

```python
#  3. Get the total amount of fares for each city type
city_total_fares = pyber_data_df.groupby(["type"]).sum()["fare"]
city_total_fares
```

```python
#  4. Get the average fare per ride for each city type. 
average_fare = city_total_fares / city_total_rides
average_fare
```

```python
# 5. Get the average fare per driver for each city type. 
average_fare_driver = city_total_fares / city_total_drivers
average_fare_driver
```

```python
#  6. Create a PyBer summary DataFrame. 
city_type_summary_df = pd.DataFrame({
             "Total Rides": city_total_rides,
             "Total Drivers": city_total_drivers,
             "Total Fares": city_total_fares,
             "Average Fare per Ride": average_fare,
             "Average Fare per Driver": average_fare_driver})
city_type_summary_df
```

```python
#  7. Cleaning up the DataFrame. Delete the index name
city_type_summary_df.index.name = None
city_type_summary_df
```

```python
#  8. Format the columns.
city_type_summary_df["Total Fares"] = city_type_summary_df["Total Fares"].map("${:,.2f}".format)
city_type_summary_df["Average Fare per Ride"] = city_type_summary_df["Average Fare per Ride"].map("${:.2f}".format)
city_type_summary_df["Average Fare per Driver"] = city_type_summary_df["Average Fare per Driver"].map("${:.2f}".format)
city_type_summary_df
```

## Deliverable 2.  Create a multiple line plot that shows the total weekly of the fares for each type of city.

```python
# 9. Using groupby() on the pyber_data_df to create a new DataFrame showing the sum of the fares 
#  for each date where the indices are the city type and date.
weekly_fares = pyber_data_df.groupby(["type", "date"]).sum()[["fare"]]
weekly_fares
```

```python
# 10. Reset the index on the DataFrame you created in #1. This is needed to use the 'pivot()' function.
weekly_fares = weekly_fares.reset_index()
weekly_fares
```

```python
# 11. Create a pivot table with the 'date' as the index, the columns ='type', and values='fare' 
# to get the total fares for each type of city by the date. 
weekly_fares_df = weekly_fares_df.pivot(index="date", columns="type", values="fare")
weekly_fares
```

```python
# 12. Create a new DataFrame from the pivot table DataFrame using loc on the given dates, '2018-01-01':'2018-04-29'.
weekly_winter_fares_df = weekly_fares.loc["2019-01-01":"2019-04-29"]
weekly_winter_fares_df
```

```python
# 13. Set the "date" index to datetime datatype. This is necessary to use the resample() method in Step 8.
weekly_winter_fares_df.index = pd.to_datetime(weekly_winter_fares_df.index)
weekly_winter_fares_df
```

```python
# 14. Check that the datatype for the index is datetime using df.info()
weekly_winter_fares_df.info()
```

```python
# 15. Create a new DataFrame using the "resample()" function by week 'W' and get the sum of the fares for each week.
total_weekly_winter_fares_df = weekly_winter_fares_df.resample("W").sum()
total_weekly_winter_fares_df
```

```python
# 16. Using the object-oriented interface method, plot the resample DataFrame using the df.plot() function. 
fig, ax = plt.subplots(figsize=(20,6))
ax.plot(total_weekly_winter_fares_df["Rural"], label="Rural")
ax.plot(total_weekly_winter_fares_df["Suburban"], label="Suburban")
ax.plot(total_weekly_winter_fares_df["Urban"], label="Urban")

ax.set_xlabel("Date")
import matplotlib.dates as mdates
ax.xaxis.set_major_locator(mdates.MonthLocator())
ax.xaxis.set_major_formatter(mdates.DateFormatter('%b'))
ax.set_ylabel("Fare ($USD)")
ax.set_title("Total Fare by City Type")
ax.legend()

# Import the style from Matplotlib.
from matplotlib import style

# Use the graph style fivethirtyeight.
style.use('fivethirtyeight')

plt.savefig("analysis/PyBer_fare_summary.png")
plt.show()
```
